```python
def moving_average(data, window_size):
    """Compute the moving average of a 1D array."""
    return np.convolve(data, np.ones(window_size), 'valid') / window_size
plt.figure(figsize=(10, 5))

profit_0 = moving_average(data['profit'][0,:], 10000)
profit_1 = moving_average(data['profit'][1,:], 10000)

plt.plot(profit_0, label='Agent 0', color='blue')
plt.plot(profit_1, label='Agent 1', color='orange')
```
