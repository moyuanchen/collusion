```python
"""
order_book_visualization.py
===========================
Simulate a singleperiod callauction (batch) market and visualise the
limitorder book (LOB) just before the auction together with the clearing
price/volume.

Quick start
-----------
$ python order_book_visualization.py    # generates a sample plot

The script does **not** require any external marketdata feed  it uses a very
simple random order generator so that the plot works straight out of the box.
Feel free to replace the generator with real depth data from an exchange API.

Requires: matplotlib, numpy (install via pip install matplotlib numpy)
"""

from __future__ import annotations

import random
from dataclasses import dataclass, field
from typing import List, Tuple, Dict, Optional

import matplotlib.pyplot as plt
import numpy as np


@dataclass
class Order:
    """A single limit order."""

    side: str  # "bid" or "ask"
    price: float  # limit price
    quantity: int  # quantity (lot size)

    def __post_init__(self) -> None:
        self.side = self.side.lower()
        if self.side not in {"bid", "ask"}:
            raise ValueError("side must be 'bid' or 'ask'")
        if self.quantity <= 0:
            raise ValueError("quantity must be positive")

    def __repr__(self) -> str:  # pragma: no cover  for nicer debugging
        return f"Order({self.side}, p={self.price}, q={self.quantity})"


@dataclass
class OrderBook:
    """A very small inmemory limit order book that only supports insertion.

    We *do not* perform continuous matching here  we simply collect orders
    until auction time. Matching is done in :py:meth:`clearing_price` by
    computing cumulative demand/supply curves.
    """

    bids: List[Order] = field(default_factory=list)
    asks: List[Order] = field(default_factory=list)

    # ---------------------------------------------------------------------
    # Public API
    # ---------------------------------------------------------------------
    def add(self, order: Order) -> None:
        """Insert a new limit order into the book."""
        (self.bids if order.side == "bid" else self.asks).append(order)

    def aggregate(self) -> Tuple[np.ndarray, np.ndarray, np.ndarray, np.ndarray]:
        """Return price levels (sorted) and cumulative quantities.

        Returns
        -------
        bid_prices : ndarray (descending)
        bid_qty    : ndarray (cumulative, same order as bid_prices)
        ask_prices : ndarray (ascending)
        ask_qty    : ndarray (cumulative, same order as ask_prices)
        """
        bid_depth: Dict[float, int] = {}
        for o in self.bids:
            bid_depth[o.price] = bid_depth.get(o.price, 0) + o.quantity
        ask_depth: Dict[float, int] = {}
        for o in self.asks:
            ask_depth[o.price] = ask_depth.get(o.price, 0) + o.quantity

        # Sort bids descending, asks ascending (market convention)
        bid_prices = np.array(sorted(bid_depth, reverse=True))
        ask_prices = np.array(sorted(ask_depth))
        bid_qty = np.cumsum([bid_depth[p] for p in bid_prices])
        ask_qty = np.cumsum([ask_depth[p] for p in ask_prices])
        return bid_prices, bid_qty, ask_prices, ask_qty

    def clearing_price(self) -> Tuple[Optional[float], Optional[int]]:
        """Find the auction clearing price (and matched volume).

        The algorithm chooses the price that *maximises* matched volume. If
        several prices tie, we pick the price that minimises the imbalance
        (surplus).  If there is still a tie we take the midpoint of the range.
        """
        bid_prices, bid_qty, ask_prices, ask_qty = self.aggregate()
        if bid_prices.size == 0 or ask_prices.size == 0:
            return None, None  # book is empty on one side

        all_prices = np.sort(np.unique(np.concatenate([bid_prices, ask_prices])))
        max_volume = -1
        best_prices: List[float] = []
        best_surplus = float("inf")

        for p in all_prices:
            demand = bid_qty[bid_prices >= p][-1] if np.any(bid_prices >= p) else 0
            supply = ask_qty[ask_prices <= p][-1] if np.any(ask_prices <= p) else 0
            traded = min(demand, supply)
            surplus = abs(demand - supply)
            if traded > max_volume or (traded == max_volume and surplus < best_surplus):
                max_volume = traded
                best_surplus = surplus
                best_prices = [p]
            elif traded == max_volume and surplus == best_surplus:
                best_prices.append(p)

        if max_volume <= 0:
            return None, None
        clearing_p = float(np.mean(best_prices))  # midpoint if range
        return clearing_p, int(max_volume)


# -------------------------------------------------------------------------
# Utility helpers
# -------------------------------------------------------------------------

def simulate_random_book(
    n_orders: int = 200,
    mid_price: float = 100.0,
    max_spread: float = 5.0,
    max_qty: int = 20,
    seed: int | None = None,
) -> OrderBook:
    """Create a random book with *n_orders* equally split between bids/asks."""
    if seed is not None:
        random.seed(seed)
    ob = OrderBook()
    for _ in range(n_orders):
        side = random.choice(["bid", "ask"])
        sign = -1 if side == "bid" else 1
        price = mid_price + sign * random.uniform(0, max_spread)
        price = round(price, 2)
        qty = random.randint(1, max_qty)
        ob.add(Order(side, price, qty))
    return ob


def plot_book(ob: OrderBook, title: str | None = None, show: bool = True) -> None:
    """Plot bid/ask depth curves and annotate clearing price."""
    bid_p, bid_q, ask_p, ask_q = ob.aggregate()
    fig, ax = plt.subplots(figsize=(8, 5))

    # Step plot (depth chart)  bids to the left, asks to the right
    ax.step(
        np.concatenate([[bid_p[0]], bid_p]),
        np.concatenate([[0], bid_q]),
        where="pre",
        label="Bids (cumulative)",
    )
    ax.step(
        np.concatenate([ask_p, [ask_p[-1]]]),
        np.concatenate([ask_q, [0]]),
        where="post",
        label="Asks (cumulative)",
    )

    cp, volume = ob.clearing_price()
    if cp is not None:
        ax.axvline(cp, linestyle="--", label=f"Clearing price = {cp:.2f}\nCleared Q = {volume}")

    ax.set_xlabel("Price")
    ax.set_ylabel("Cumulative quantity")
    ax.set_title(title or "Callauction limit order book")
    ax.legend()
    ax.grid(True, linewidth=0.3)
    plt.tight_layout()
    plt.xlim(98,102)
    plt.ylim(0, 500)
    if show:
        plt.show()


# -------------------------------------------------------------------------
# CLI entry point
# -------------------------------------------------------------------------

if __name__ == "__main__":
    book = simulate_random_book(seed=42)
    plot_book(book)
    
```

```python
def _grid_linspace(low: float, high: float, n: int) -> np.ndarray:
    """Helper that always returns float64 equallyspaced grid."""
    return np.linspace(low, high, n, dtype=np.float64)


def build_discretisations(cfg, factor = 1) -> Tuple[np.ndarray, np.ndarray, np.ndarray]:
    """Return (p_disc, v_disc, x_disc) as numpy 1D arrays."""
    # ----- v grid (inverseCDF of standard normal) -----
    from math import sqrt
    from scipy.stats import norm  # lightweight import
    probs = (2 * np.arange(1, cfg.Nv + 1) - 1) / (2 * cfg.Nv)
    v_disc = cfg.v_bar + cfg.sigma_v * norm.ppf(probs)

    # ----- x grid using chiN / chiM -----
    span_x = abs(cfg.chi_N - cfg.chi_M) * factor
    low_x = -max(cfg.chi_N, cfg.chi_M) - cfg.iota * span_x
    high_x = max(cfg.chi_N, cfg.chi_M) + cfg.iota * span_x
    x_disc = _grid_linspace(low_x, high_x, cfg.Nx)

    # ----- p grid (use widest lambda) -----
    lam_max = max(cfg.lambda_N, cfg.lambda_M)
    ph = cfg.v_bar + lam_max * (cfg.I * max(cfg.chi_N, cfg.chi_M) + cfg.sigma_u * 1.96)
    pl = cfg.v_bar - lam_max * (cfg.I * max(cfg.chi_N, cfg.chi_M) + cfg.sigma_u * 1.96)
    span_p = ph - pl
    p_disc = _grid_linspace(pl - cfg.iota * span_p, ph + cfg.iota * span_p, cfg.Np)

    return p_disc, v_disc, x_disc

p_disc, v_disc, x_disc = build_discretisations(cfg)
```

```python
def initialise_Q_batch(cfg, p_disc, v_disc, x_disc, B: int):
    """
    Build a (B, I, Np, Nv, Nx) tensor: one Q-table per batch path.
    Returns
    -------
    np.ndarray  float32  shape (B, I, Np, Nv, Nx)
    """
    I, Np, Nv, Nx = cfg.I, cfg.Np, cfg.Nv, cfg.Nx
    v_bar, lamN, x_n, rho = cfg.v_bar, cfg.lambda_N, cfg.chi_N, cfg.rho

    # ---- constants over v ----
    Sx      = x_disc.sum()                                  #  x_i
    base_v  = v_disc - v_bar - lamN * x_n                   # (Nv,)
    const_v = (Nx * base_v - lamN * (I - 1) * Sx) / ((1 - rho) * Nx)  # (Nv,)

    # ---- core 4-D slab (1, I, Np, Nv, Nx) ----
    core = const_v[None, None, None, :, None] * x_disc[None, None, None, None, :]
    #              ^        ^     ^               broadcast over I, Np, Nx

    # ---- broadcast to include batch ----
    return np.broadcast_to(core, (B, I, Np, Nv, Nx)).astype(np.float32)

# def initialise_Q(cfg, p_disc, v_disc, x_disc):
#     """
#     Build a (I, Np, Nv, Nx) Q-table in one vectorised shot, using
#     parameters directly from the `cfg` object.
#     Returns
#     -------
#     np.ndarray   float32   shape (I, Np, Nv, Nx)
#     """
#     # unpack once for readability
#     I, Np, Nv, Nx = cfg.I, cfg.Np, cfg.Nv, cfg.Nx
#     v_bar   = cfg.v_bar
#     lamN    = cfg.lambda_N
#     x_n     = cfg.chi_N
#     rho     = cfg.rho

#     # ------------------------------------------------------------------
#     # constants that depend only on v-index
#     # ------------------------------------------------------------------
#     Sx        = x_disc.sum()                                #  x_i
#     base_v    = v_disc - v_bar - lamN * x_n                 # (Nv,)
#     const_v   = (Nx * base_v - lamN * (I - 1) * Sx)         # (Nv,)
#     const_v  /= (1.0 - rho) * Nx                            # (Nv,)

#     # ------------------------------------------------------------------
#     # broadcast into full Q-table   Q[i,p,v,x] = const_v[v] * x_disc[x]
#     # ------------------------------------------------------------------
#     Q = const_v[None, None, :, None] * x_disc[None, None, None, :]     # (1,1,Nv,Nx)
#     return np.broadcast_to(Q, (I, Np, Nv, Nx)).astype(np.float32)


# # ---- usage -----------------------------------------------------------

```

```python
class Ring:
    """Circular 1-D buffer that returns a *view* in chronological order."""
    __slots__ = ("buf", "idx")

    def __init__(self, size, dtype=np.float64):
        self.buf = np.zeros(size, dtype=dtype)
        self.idx = 0                 # write pointer

    def add(self, value):
        self.buf[self.idx] = value
        self.idx = (self.idx + 1) % self.buf.size

    def view(self):
        """Return a 1-D **view**: newest element is at index 1."""
        i = self.idx
        return np.concatenate((self.buf[i:], self.buf[:i]), axis=0)

class OnlineOLS:
    """2-var OLS for y =  +   x   with sliding window."""
    __slots__ = ("ring_x", "ring_y", "Sxx", "Sx", "Sxy", "Sy", "n")

    def __init__(self, size):
        self.ring_x = Ring(size)
        self.ring_y = Ring(size)
        self.Sxx = self.Sx = self.Sxy = self.Sy = 0.0
        self.n = 0                 # samples stored

    def add(self, x, y):
        # remove oldest if buffer full
        if self.n == self.ring_x.buf.size:
            x_old = self.ring_x.buf[self.ring_x.idx]
            y_old = self.ring_y.buf[self.ring_y.idx]
            self.Sxx -= x_old * x_old
            self.Sx  -= x_old
            self.Sxy -= x_old * y_old
            self.Sy  -= y_old
        else:
            self.n += 1

        # add new
        self.ring_x.add(x)
        self.ring_y.add(y)
        self.Sxx += x * x
        self.Sx  += x
        self.Sxy += x * y
        self.Sy  += y

    def coef(self):
        # Solve 22 system analytically
        det = self.n * self.Sxx - self.Sx**2
        if det == 0:
            return 0.0, 0.0
        1 = (self.n * self.Sxy - self.Sx * self.Sy) / det
        0 = (self.Sy - 1 * self.Sx) / self.n
        return 1, 0
class AdaptiveMarketMaker:
    """
    Vectorised market maker with O(1) update; batched over B paths if yt is array.
    """
    def __init__(self, cfg):
        self.theta = cfg.theta
        m = cfg.Tm
        # OLS trackers
        self.ols_zp = OnlineOLS(m)   # xi_1 from z  p
        self.ols_yv = OnlineOLS(m)   # ,  from y  v

    def determine_price(self, yt):
        xi_1, _      = self.ols_zp.coef()
        gamma_1, g0  = self.ols_yv.coef()
        lambda_      = (xi_1 + self.theta * gamma_1) / (xi_1**2 + self.theta)
        return g0 + lambda_ * yt

    def update(self, vt, pt, zt, yt):
        """vt, pt, zt, yt can be scalars **or 1-D vector of batch size B**."""
        # flatten in case of numpy 0-D
        vt = np.asarray(vt).ravel()
        pt = np.asarray(pt).ravel()
        zt = np.asarray(zt).ravel()
        yt = np.asarray(yt).ravel()
        # loop only over batch dimension (tiny)
        for x, y in zip(zt, pt):
            self.ols_zp.add(x, y)
        for x, y in zip(yt, vt):
            self.ols_yv.add(x, y)
```

```python
rng     = np.random.default_rng(123)
T       = 5000
B       = 1000
noise   = rng.normal(0, cfg.sigma_u, (B,T))
v_path  = rng.integers(0, cfg.Nv, (B, T))

p_idx   = rng.integers(0, cfg.Np, (B,))   # start price index
v_idx   = v_path[:, 0]                     # (B,)  initial value indices

# ---------- market-maker ----------
mm = AdaptiveMarketMaker(cfg)              # one shared MM

# ---------- discretisations ----------
p_disc, v_disc, x_disc = build_discretisations(cfg)

# ---------- per-path Q, visit, last-opt ----------
Q_table     = initialise_Q_batch(cfg, p_disc, v_disc, x_disc, B)  # (B,I,Np,Nv,Nx)
visit_count = np.zeros((B, cfg.I, cfg.Np, cfg.Nv),     dtype=np.int32)
last_opt    = -np.ones((B, cfg.I, cfg.Np, cfg.Nv),     dtype=np.int16)
conv_ctr    = np.zeros((B, cfg.I),             dtype=np.int32)

# helper index tensors
batch_ix = np.arange(B)[:, None]           # shape (B,1)
agent_ix = np.arange(cfg.I)[None, :]           # shape (1,I)

# ---------- initial memory (greedy value) ----------
greedy  = Q_table[batch_ix, agent_ix, p_idx[:,None], v_idx[:,None]].argmax(-1)   # (B,I)
memory  = np.take_along_axis(
            Q_table[batch_ix, agent_ix, p_idx[:,None], v_idx[:,None]],
            greedy[..., None], axis=-1
         )[..., 0]                                                              # (B,I)

# ---------- main loop ----------
for t in range(T - 1):
    u_t = noise[:, t]                             # (B,)

    # -greedy
    eps      = np.exp(-cfg.beta * visit_count[batch_ix, agent_ix, p_idx[:,None], v_idx[:,None]])
    visit_count[batch_ix, agent_ix, p_idx[:,None], v_idx[:,None]] += 1

    explore  = rng.random((B, I)) < eps           # (B,I)
    greedy   = Q_table[batch_ix, agent_ix, p_idx[:,None], v_idx[:,None]].argmax(-1)
    randomA  = rng.integers(0, cfg.Nx, (B, cfg.I))
    action   = np.where(explore, randomA, greedy)           # (B,I)

    # trades & market maker
    y_sum = x_disc[action].sum(axis=1) + u_t               # (B,)
    p_val = mm.determine_price(y_sum)                      # float or (B,) if vectorised MM
    z_val = cfg.xi * (p_val - cfg.v_bar)
    mm.update(v_disc[v_idx], p_val, z_val, y_sum)

    # rewards & Bellman
    profit   = x_disc[action] * (v_disc[v_idx,None] - p_val[:,None])    # (B,I)
    best     = Q_table[batch_ix, agent_ix, p_idx[:,None], v_idx[:,None]].max(-1)
    bellman  = (1-cfg.alpha)*memory + cfg.alpha*(profit + cfg.rho*best)

    # write back
    Q_table[batch_ix, agent_ix, p_idx[:,None], v_idx[:,None], action] = bellman
    memory = bellman

    # convergence counter
    changed = greedy != last_opt[batch_ix, agent_ix, p_idx[:,None], v_idx[:,None]]
    reset   = changed.any(axis=1)
    conv_ctr[reset]  = 0
    conv_ctr[~reset] += 1
    last_opt[batch_ix, agent_ix, p_idx[:,None], v_idx[:,None]] = greedy

    # advance indices for next step
    v_idx = v_path[:, t+1]                                              # (B,)
    p_idx = np.abs(p_disc[:,None] - p_val).argmin(axis=0)               # (B,)
```
